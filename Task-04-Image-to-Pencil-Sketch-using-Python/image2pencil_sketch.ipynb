## `Step-1:` Importing Libraries.

```python
import cv2
import matplotlib.pyplot as plt
```

## `Step-2:` Loading Original Image.

```python
img = cv2.imread("./images/shrinkhala.jpg")
img[:1]
```

### OpenCV uses BGR image format. 

#### So, when we read an image using cv2. imread() it interprets in BGR format by default. We can use cvtColor() method to convert a BGR image to RGB and vice-versa

`cv2. imread()` method returns a BGR values as numpy.ndarray (NumPy N-dimensional array) if the loading of the image is successful. Otherwise, It returns an empty matrix ( Mat::data==NULL ) if the image cannot be read because of any reason (like missing file, improper permissions, unsupported or invalid format). 

`imread` returns the image data in the array A . If the file contains a grayscale image, A is a two-dimensional (M-by-N) array. If the file contains a color image, A is a three-dimensional (M-by-N-by-3) array. The class of the returned array depends on the data type used by the file format.

### Converting into RGB image.

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
rgb_img[:1]
```

You can see `B` columns have values of `R` columns and vice-versa.

```python
fig ,(ax0, ax1) = plt.subplots(nrows=1,
                               ncols=2,
                               figsize= (10,6))

ax0.axis('off')
ax1.axis('off')

ax0.imshow(img)
ax0.set(title="BGR Image");
ax0.title.set_color("purple")

ax1.imshow(rgb_img)
ax1.set(title="RGB Image")
ax1.title.set_color("purple")


# Add a title to figure

fig.suptitle("Conversion from Original to Pencil-Sketch",
            fontsize =16,
            fontweight="bold",
            color="orange");
```

## `Step-3:` Converting BGR into GrayScale Image.

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  #ConvertColor--cvtColor

## Visualizing image using matplotlib

plt.imshow(gray_img)
plt.axis('off')
plt.title("GrayScale Image").set_color("purple")
```

## `Step-4:` Inverting Gray Scaled Image.

Two ways to invert an image.

  a) Subtracting from 255.
    
  b) Using bitwise NOT operator.

```python
# Subtracting from 255.
    # invert_img = 255 - gray_img
    
invert_img = cv2.bitwise_not(gray_img) # using bitwise NOT operator

## Visualizing image using matplotlib

plt.imshow(invert_img)
plt.axis('off')
plt.title("Inverted GrayScale Image").set_color("purple")
```

## `Step-5:` Bluring Inverted Image.

```python
blur_img = cv2.GaussianBlur(invert_img, (21, 21), 0) # Play with different values.

## Visualizing image using matplotlib

plt.imshow(blur_img)
plt.axis('off')
plt.title("Blurred Image").set_color("purple")
```

## `Step-6:` Inverting Blurred Image.

```python
invert_blur_img = cv2.bitwise_not(blur_img)

## Visualizing image using matplotlib

plt.imshow(invert_blur_img)
plt.axis('off')
plt.title("Inverted Blurred Image").set_color("purple")
```

## `Step-6:` Generating Pencil Sketch Image.

```python
sketch_img = cv2.divide(gray_img, invert_blur_img, scale=256.0)

## Visualizing image using matplotlib

plt.imshow(sketch_img)
plt.axis('off')
plt.title("Pencil Sketched Image").set_color("purple")
```

This will not visualize clearly as pencil sketch using matplotlib because of yellow background. Let's visualize it after saving to local diretory and loading from there.

## `Step-8:` Saving and loading Image.

```python
cv2.imwrite('./images/sketch-img.jpg', sketch_img)


# Loaded image will be in BGR format so converting it into RGB and displaying.

output_bgr_img = cv2.imread('./images/sketch-img.jpg')
output_img = cv2.cvtColor(output_bgr_img, cv2.COLOR_BGR2RGB)

fig ,(ax0, ax1) = plt.subplots(nrows=1,ncols=2, figsize= (10,6))
ax0.axis('off')
ax1.axis('off')

ax0.imshow(rgb_img)
ax0.set(title="Original Image");
ax0.title.set_color("purple")

ax1.imshow(output_img)
ax1.set(title="Sketched Image")
ax1.title.set_color("purple")


# Add a title to figure

fig.suptitle("Original Image vs Sketched Image",
            fontsize =16,
            fontweight="bold",
            color="orange");
```

## `Step-9:` Visualization  in a Single Frame

```python
import cv2
import matplotlib.pyplot as plt

    
img = cv2.imread("./images/shrinkhala.jpg")

gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
invert_gray_img = cv2.bitwise_not(gray_img)
blur_img = cv2.GaussianBlur(invert_gray_img, (21, 21), 0) 
invert_blur_img = cv2.bitwise_not(blur_img)  
sketch_img = cv2.divide(gray_img, invert_blur_img, scale= 256.0) 

cv2.imwrite("./images/sketch-image.jpg", sketch_img)

output_bgr_img = cv2.imread('./images/sketch-image.jpg')
output_img = cv2.cvtColor(output_bgr_img, cv2.COLOR_BGR2RGB)

fig ,((ax0, ax1, ax2), (ax3, ax4, ax5)) = plt.subplots(nrows=2,
                                                       ncols=3,
                                                       figsize= (10,10))
ax0.axis('off')
ax0.imshow(img)
ax0.set(title="BGR image")

ax1.axis('off')
ax1.imshow(gray_img)
ax1.set(title="Gray image")

ax2.axis('off')
ax2.imshow(invert_gray_img)
ax2.set(title="Inverted Gray image")

ax3.axis('off')
ax3.imshow(blur_img)
ax3.set(title="Blurred image")

ax4.axis('off')
ax4.imshow(invert_blur_img)
ax4.set(title="Inverted Blurred image")

ax5.axis('off')
ax5.imshow(output_img)
ax5.set(title="Sketch image")



# Add a title to figure

fig.suptitle("Conversion from Original(BRG) to Pencil-Sketch",
            fontsize =16,
            fontweight="bold",
            color="orange");
```
